```python
import pandas as pd

import random
import time

from bs4 import BeautifulSoup
import pandas as pd
import requests
from fake_useragent import UserAgent
from requests import request
from requests.compat import urljoin, urlparse
from requests.exceptions import HTTPError
from urllib.robotparser import RobotFileParser
from requests import Session
import re
from fake_useragent import UserAgent
```

```python
data = pd.read_csv('../data/ted_talks.csv')
data = pd.DataFrame(data)
data['levels']=1 # 나중에 난이도 생기면 수정바랍니다
```

```python
# 처음 코드 돌리는 거면 돌려주세요. 그 외에는 돌리시면 안되요
data['image']=None
```

```python
# 중간에 Attribute Error뜨는건 서버 렉, 혹은 본인 네트워크때문...일단 try, except문으로 처리했으니 다른 에러뜨면 수정해주세요
# UserAgent쓰는 이유: 혹시 서버렉 일으켰다고 테드측에서 혼낼까봐...빼면 좀 더 빨라지긴 할겁니다..
urls=list(data['url'])
for i in range(0,len(urls)):
    try:
        ua = UserAgent()
        headers = {'User-Agent' : ua.random}
        url=urls[i]
        req = requests.get(url,headers=headers)
        html = req.text 
        soup = BeautifulSoup(html, 'html.parser')
        image=soup.find(attrs={'property': 'twitter:image'}).attrs['content']
        print(str(i),'번째')
        print(url)
        print(image)
        data['image'][i]=image
    except Exception as e:
#        print(e)
        url=urls[i]
        req = requests.get(url,headers=headers)
        html = req.text 
        soup = BeautifulSoup(html, 'html.parser')
        image=soup.find(attrs={'property': 'twitter:image'}).attrs['content']
        print(str(i),'번째')
        print(url)
        print(image)
        data['image'][i]=image
```
